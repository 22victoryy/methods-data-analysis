```r


## Simulation ##
set.seed(1002656486)
beta0 <- rnorm(n = 1, mean = 0, sd = 1) ## The population beta0 
beta1 <- runif(n = 1, min = 1, max = 3)  ## The population beta1
sig2 <- rchisq(n = 1, df = 1) ## The error variance sigma^2

## Multiple simulation may require loops ##
nsample <- 5  ## Sample size
n.sim <- 10 ## The number of simulations
sigX <- 0.2 ## The variances of X
## Simulate the predictor variable ##
X <- rnorm(n = nsample, mean = 0, sd = sqrt(sigX))


##########

### Simulate ##
b0 <- vector() ## saves the sample estimates of beta0
b1 <- vector() ## saves the sample estimates of beta1
sig2hat <- vector()  ## saves the sample estimates of sigma^2
sigy <- vector()  ## Saves the estimate of error variance
for(i in 1:n.sim){
Y <- beta0 + beta1*X + rnorm(n = nsample, mean = 0, sd = sqrt(sig2))
model <- lm(Y ~ X)
b0[i] <- coef(model)[1]
b1[i] <- coef(model)[2]
sigy[i] <- summary(model)$sigma^2
}

mean(b0); mean(b1); mean(sigy)


beta0; beta1; sig2
par(mfrow = c(2,2))
hist(b0)
hist(b1)
hist(sigy)
```
